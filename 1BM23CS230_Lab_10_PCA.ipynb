```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("heart.csv")

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
le=LabelEncoder()

for col in df.columns:
  if df[col].dtype=='object':
    df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

```python
X=df.drop('HeartDisease', axis=1)
Y=df['HeartDisease']

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2)

scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
lr=LogisticRegression()
lr.fit(X_train, Y_train)

Y_pred=lr.predict(X_test)
print("LR Accuracy: ", accuracy_score(Y_test, Y_pred))
```

```python
svm=SVC()
svm.fit(X_train, Y_train)

Y_pred=svm.predict(X_test)
print("SVM Accuracy: ",accuracy_score(Y_test, Y_pred))
```

```python
rf=RandomForestClassifier()
rf.fit(X_train, Y_train)

Y_pred=rf.predict(X_test)
print("RF Accuracy: ", accuracy_score(Y_test, Y_pred))
```

```python
pca=PCA(n_components=5)
```

```python
X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
lr.fit(X_train_pca, Y_train)

Y_pred=lr.predict(X_test_pca)
print("LR Accuracy after PCA: ", accuracy_score(Y_test, Y_pred))
```

```python
svm.fit(X_train_pca, Y_train)

Y_pred=svm.predict(X_test_pca)
print("SVM Accuracy after PCA: ", accuracy_score(Y_test, Y_pred))
```

```python
rf.fit(X_train_pca, Y_train)

Y_pred=rf.predict(X_test_pca)
print("RF Accuracy after PCA: ", accuracy_score(Y_test, Y_pred))
```
