```python
import pandas as pd

df=pd.read_csv("income.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
X=df.drop("income_level", axis=1)
Y=df["income_level"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_10=AdaBoostClassifier(n_estimators=10, random_state=42)
ada_10.fit(X_train, Y_train)

Y_pred_10=ada_10.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy=accuracy_score(Y_test, Y_pred_10)
cm=confusion_matrix(Y_test, Y_pred_10)

print("Accuracy for 10 estimators: ", accuracy)
print("Confusion Matrix for 10 estimators: ", cm)
```

```python
best_score=0
best_n_estimators=0



for est in [10, 20, 50, 100, 150, 160, 180, 200]:
  model=AdaBoostClassifier(n_estimators=est, random_state=42)
  model.fit(X_train, Y_train)
  Y_pred=model.predict(X_test)
  accuracy=accuracy_score(Y_test, Y_pred)

  if (accuracy > best_score):
    best_score=accuracy
    best_n_estimators=est


print("Best Accuracy for ", best_n_estimators," is found to be : ", best_score)

model=AdaBoostClassifier(n_estimators=best_n_estimators, random_state=42)
model.fit(X_train, Y_train)
Y_pred=model.predict(X_test)
cm2=confusion_matrix(Y_test, Y_pred)

print("Confusion Matrix for  ", best_n_estimators," estimators: ", cm2)
```
