```python
df = pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)

"""Using RBF kernel"""

from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
len(X_train)
```

```python
len(X_test)
```

```python
rbf_model.fit(X_train, y_train)

rbf_model.score(X_test,y_test)
```

```python
"""Using Linear kernel"""

linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)

linear_model.score(X_test,y_test)
```

SVM - Iris

```python
import pandas as pd

df1=pd.read_excel("iris (1).xlsx")
df1.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix

X=df1.drop('species', axis=1)
Y=df1['species']

X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)

```

```python
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)

model1=SVC(kernel='linear')
model1.fit(X_train, Y_train)

y_pred1=model1.predict(X_test)
accuracy1=accuracy_score(Y_test, y_pred1)
cm1=confusion_matrix(Y_test, y_pred1)

print("Accuracy of Linear Kernel is : ", accuracy1)
print("\nConfusion Matrix for Linear Kernel : \n", cm1)
```

```python
model2=SVC(kernel='rbf')
model2.fit(X_train, Y_train)

y_pred2=model2.predict(X_test)
accuracy2=accuracy_score(Y_test, y_pred2)
cm2=confusion_matrix(Y_test, y_pred2)

print("Accuracy of RBF Kernel is : ", accuracy2)
print("\nConfusion Matrix for RBF Kernel : \n", cm2)
```
